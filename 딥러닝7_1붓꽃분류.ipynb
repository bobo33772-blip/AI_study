```python
import pandas as pd
url='https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
iris = pd.read_csv(url)
iris.info()
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
iris.describe()
```

```python
# 입력, 타겟 분리
X = iris.drop(columns='species')
y = iris['species']
```

```python
# 품종을 숫자로 변환
# y_encode = y.map(
#     {'setosa':0, 'versicolor':1, 'virginica':2}
# )
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encode = le.fit_transform(y)
y_encode
# 원핫인코딩 - X에 적용
# y_encode = pd.get_dummies(y,dtype=int)
# y_encode.head()
```

```python
# StandardScaler 적용
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
# train_test_split()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_encode,
    test_size=0.2,
    random_state=42,
    stratify=y_encode
)
```

```python
import keras
# Input(), Dense(), Sequential()
# shape=(4,) -> 입력시 4개씩 전달됨
# (sepal_length	sepal_width	petal_length	petal_width)
inputs = keras.layers.Input(shape=(4,))
# 3 -> 3가지 값
# softmax -> 각각의 확률로 반환
dense = keras.layers.Dense(3, activation='softmax')
# Sequential()
model = keras.Sequential([inputs, dense])
```

```python
# compile()
model.compile(
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# fit()
model.fit(X_train, y_train, epochs=5)
```

```python
# predict()
print('입력', X_train[0:1])
print('타겟', y_train[0:1])
model.predict( X_train[0:1] )
pred = model.predict(X_train[0:1])
print('pred',pred)

import numpy as np
print('예상타겟',np.argmax(pred))
```

```python
# evaluate()
score_tr = model.evaluate(X_train, y_train)
print('score_tr', score_tr)
score_te = model.evaluate(X_train, y_train)
print('score_te', score_te)
```
